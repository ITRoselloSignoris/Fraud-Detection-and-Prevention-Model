## 1- Cargamos los datos

```python
import pandas as pd

data = pd.read_csv("../datasets/processed_data/ohe_customer_dataset.csv", sep = ",")
data
```

```python
from funpymodeling.exploratory import status

status(data)
```

En este caso, vamos a analizar solo algunas columnas que nos parecen relevantes.

```python
filter_data = data[["orderState_fulfilled", "orderState_pending", "orderState_failed", "transactionFailed", "fraudulent", "customerIPAddressSimplified_digits_and_letters"]].copy()
```

```python
status(filter_data)
```

## 2- Clustering con K-Means

### 2.1- Normalización de datos 

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
std_scaler.fit(filter_data)
x_data_norm = std_scaler.transform(filter_data)

pd.DataFrame(x_data_norm).describe().T
```

### 2.2- Creación del modelo

```python
from sklearn.cluster import KMeans

cluster_model = KMeans(n_clusters = 3,
                       init = "k-means++",
                       max_iter = 100,
                       random_state = 0,
                       n_init = "auto")

cluster_model.fit(x_data_norm)
```

```python
pred_cluster = cluster_model.predict(x_data_norm)
pred_cluster
```

### 2.3- Cantidad de clusters

```python
cluster_model
```

```python
from copy import deepcopy
```

```python
from yellowbrick.cluster.elbow import kelbow_visualizer

cluster_model2 = deepcopy(cluster_model)

kelbow_visualizer(cluster_model2, x_data_norm, k = (2,30), locate_elbow = True)
```

Podemos observar que la cantidad óptima de clusters es __9__.

### 2.4- Gráfico de coordenadas

```python
cluster_var = "cluster"
x_data_cluster = filter_data.copy()
x_data_cluster[cluster_var] = cluster_model.predict(x_data_norm)
```

```python
x_data_cluster.head(5)
```

```python
from funpymodeling.model_validation import coord_plot
```

```python
d_orig, d_transf = coord_plot(x_data_cluster, cluster_var)
```

**Observaciones Clusters K-Means:**

- En el cluster __0__ podemos ver que corresponde a cuando es un _warning_. Para este cluster podemos notar que el estado de orden es similar al estado de orden del cluster __2__, pero luego para la transacción y la dirección IP es similar al cluster __1__. Se puede concluir que posee valores de cuando es fraude y de cuando no lo es, por eso mismo es un _warning_.

- En el cluster __1__ podemos ver que corresponde a cuando _es fraude_. La conclusión más notable es que todos los valores de las categorías observamos, son opuestos al del cluster 2, no coincide ninguno.

- En el cluster __2__ podemos ver que corresponde a cuando _no es fraude_. Otra detalle interesante es que a diferencia de los otros 2 clusters, la dirección IP del cliente se compone de dígitos y letras, y la transacción falló.

## 3- Clustering con HDBSCAN

### 3.1- Creación del modelo

```python
import hdbscan
```

```python
model_hdb = hdbscan.HDBSCAN().fit(filter_data)
```

```python
from funpymodeling.exploratory import freq_tbl

freq_tbl(model_hdb.labels_.astype("str"))
```

### 3.2- Tuning/regularización de HDBSCAN para menos clusters

```python
model_hdb2 = hdbscan.HDBSCAN(min_cluster_size = 15).fit(filter_data)
```

```python
freq_tbl(model_hdb2.labels_.astype("str"))
```

### 3.3- Análisis del modelo cluster (profiling)

```python
cluster_var = "cluster"
x_data_cluster = filter_data.copy()
x_data_cluster[cluster_var] = model_hdb2.labels_
```

```python
x_data_cluster.head()
```

```python
x_data_cluster = x_data_cluster[x_data_cluster["cluster"] != -1]
```

```python
d_orig, d_transf = coord_plot(x_data_cluster, cluster_var)
```

**Observaciones Clusters HDBSCAN:**

- Un detalle bastante interesante, es que ningún cluster que pertenezca a _warning_, tiene la dirección IP con dígitos y letras, siendo estos clusters la mayoría, específicamente 5 clusters.

- Como el cluster 6 es el único que pertenece a  _fraude_, voy a mencionar ciertos detalles de sus otras categorías.   
    - La transacción no falló.
    - El estado de la orden se cumplió.
    - No tiene un dirección IP con dígitos y letras.
    
- Hay 1 cluster donde el estado de la orden es falso para las 3 columnas, cluster __1__, por lo tanto, tenemos este cluster que no nos estarían diciendo nada del estado de la orden.

- Gran parte de los estados de orden son cumplidos, excepto para el cluster mencionado anteriormente, para el cluster __0__, donde podemos ver que el estado de orden es fallido, y para el cluster __2__, donde podemos ver que el estado de orden es pendiente.

- Para el gráfico de coordenadas de K-Means, pudimos notar que el único cluster donde la dirección IP tenía dígitos y letras era el de _no fraude_, y si observamos el cluster __5__ podemos notar una similitud con respecto a esas condiciones mencionadas.

### 4- Cálculo de Silhouette

```python
from sklearn.metrics import silhouette_score, silhouette_samples

sil_avg = silhouette_score(filter_data, model_hdb2.labels_)
```

```python
sil_avg
```

```python
sample_silhouette_values = silhouette_samples(filter_data, model_hdb2.labels_)
```

```python
sample_silhouette_values
```

```python
res_sil = pd.DataFrame({"sil_value": sample_silhouette_values, "cluster": model_hdb2.labels_})
```

```python
import seaborn as sns
```

```python
sns.kdeplot(data=res_sil, x="cluster", y="sil_value", fill=True)
```

```python
sns.violinplot(data=res_sil, x="cluster", y="sil_value", fill=True)
```

```python
sns.histplot(data=res_sil, x="cluster", y="sil_value", fill=True)
```

```python
res_sil.groupby("cluster").mean("sil_value")
```

__Observaciones:__

- Podemos ver que el sil_value es altísimo, por lo tanto, podríamos decir que en todos se notan las diferencias entre sí, es decir, los puntos de cada cluster no están cerca de los puntos de otros clusters vecinos.
