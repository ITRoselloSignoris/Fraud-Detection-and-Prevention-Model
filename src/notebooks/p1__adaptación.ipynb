### 4.2 - Customer Email

Los tipos de mails más comúnes dentro de los clientes son, `hotmail`, `gmail` y `yahoo`.

Como muchas veces, ciertos clientes con intenciones maliciosas usan otro tipo de mail, como del tipo `@1jcfcxs7`, los podemos clasificar como `weird`.

```python
email_domains = []
mails = []
popular_providers = ["yahoo", "gmail", "hotmail"]

for email in data["customerEmail"]:
    try:
        aux = email.split("@")[1]
        domain = aux.split(".")[1]
        mail = aux.split(".")[0]

        if mail in popular_providers:
            mails.append(mail)
        else:
            mails.append("other")
        email_domains.append(domain)
    except:
        email_domains.append("weird")
        mails.append("weird")
```

```python
data["emailDomain"] = email_domains
data["emailProvider"] = mails
```

```python
data.drop(["customerEmail"],inplace=True, axis=1)
```

Debido a que hay IPs más extensas con más de 15 carácteres y además, esos caracteres no son solo números sino que también letras.

Por eso mismo, creamos un nueva columna `CustomerIPAddressSimplified` con 2 valores: `only_numeric_digits` y `digits_and_letters` para cada tipo de dirección IP según su longitud y tipo.

```python
IP_adresses = []

for adress in data["customerIPAddress"]:
    aux_adress = str(adress)

    if len(aux_adress) > 15:
        IP_adresses.append("digits_and_letters")
    else:
        IP_adresses.append("only_numeric_digits")
```

```python
data["customerIPAddressSimplified"] = IP_adresses
```

```python
data.drop(["customerIPAddress"],axis=1,inplace=True)
```

### 4.4 - customerBillingAddress and orderShippingAdress

```python
data["customerBillingAddress"].head(5)
```

```python
data["orderShippingAddress"].head(5)
```

Vamos a buscar quedarnos unicamente con la ciudad, _Por ejemplo_: Del string `898 Henry Ports\nNew Keithview, CA 95893-2497` nos quedamos con `CA`.

```python
def extract_city_name(name_column):
    city_list = []

    for adress in data[name_column]:
        try:
            aux = adress.split("\n")[1]
            city_with_number = aux.split(", ")[1]
            city = city_with_number.split(" ")[0]

            city_list.append(city)

        except:
            city_list.append("unknown")
    
    return city_list
```

```python
data["customerBillingAddress"] = extract_city_name("customerBillingAddress")
data["orderShippingAddress"] = extract_city_name("orderShippingAddress")
```

```python
data.head(5)
```

Ahora vamos a evaluar si las 2 ciudades que encontramos en `customerBillingAddress` y `orderShippingAddress` son las mismas, ya que, en el caso de no ser la misma, podría indicar la posibilidad de que sea fraude.

De esa manera, creamos una nueva columna `sameCity`, donde va a tener como valores, **yes**, **no** y **unknown**

```python
same_city = []

for index in range(len(data)):
    billing_address = data["customerBillingAddress"][index]
    order_address = data["orderShippingAddress"][index]

    response = "unknown"
    if billing_address != response and order_address != response:
        if billing_address == order_address:
            response = "yes"
        else:
            response = "no"

    same_city.append(response)
```

```python
data["sameCity"] = same_city
```

```python
data.drop(["customerBillingAddress","orderShippingAddress"],axis=1,inplace=True)
```

```python
data.head(5)
```
